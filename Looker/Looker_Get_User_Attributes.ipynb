```python
import looker_sdk
import pandas as pd
sdk = looker_sdk.init31()  # or init40() for v4.0 API
# my_user = sdk.me()
# # output can be treated like a dictionary
# print(my_user["first_name"])
# # or a model instance (User in this case)
# print(my_user.first_name)
# print(type(my_user))
all_user_id_list = sdk.all_users(fields="id", page=10)
all_user_id_list = [x.id for x in all_user_id_list]
all_user_id_list = [str(x) for x in all_user_id_list]
#all_user_id_list =  ",".join([x for x in all_user_id_list])

name_list=[]
value_list =[]
source_list = []

#all_user_id_list=id=["220","179"]

for id in all_user_id_list:
  #x=y=z=0
  user_attribute_user_values = sdk.user_attribute_user_values(id,user_attribute_ids="3,6,15,16")
  name_list.append([x.name for x in user_attribute_user_values])
  value_list.append([x.value for x in user_attribute_user_values])
  #source_list.append([x.source for x in user_attribute_user_values])


# name_list =  [x.name for x in user_attribute_user_values]
# user_id_list = [x.user_id for x in user_attribute_user_values]
# value_list = [x.value for x in user_attribute_user_values]
# df = pd.DataFrame(
#       {'name': name_list,
#      'user_id': user_id_list,
#      'value': value_list
#     })
# print(df.name)#[df["user_id"] == 179])

df = pd.DataFrame(value_list, columns=name_list[0])
```

```python
df.to_csv('/content/drive/MyDrive/UserAttributes.csv', index=False,header=True)
```
